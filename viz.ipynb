## Categorical distribution

```python
top = Table.read_table('data/top_movies.csv')
top
```

```python
# we can format better the numbers
top.set_format([2,3], NumberFormatter)
```

```python
top
```

```python
#select
movies_and_studios = top.select('Title', 'Studio')
movies_and_studios
```

```python
#group by studio
movies_and_studios.group('Studio')
```

```python
#sort and draw chart
movies_and_studios.group('Studio').sort('count', descending=True).barh(0)
```

```python
#I wanted to create a new table just with title and total gross in millions, here is the way
millions = top.select(0).with_columns('Gross - Mil Dollars', np.round(top.column(3)/1000000,2))
millions
```

```python
# histogram
millions.hist('Gross - Mil Dollars', unit='$m')
```

```python
millions.hist('Gross - Mil Dollars', bins=np.arange(300,2001,100), unit='$m')
```
