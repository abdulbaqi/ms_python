## Categorical distribution

```python
top = Table.read_table('data/top_movies.csv')
top
```

```python
# we can format better the numbers
top.set_format([2,3], NumberFormatter)
```

```python
top
```

```python
#select
movies_and_studios = top.select('Title', 'Studio')
movies_and_studios
```

```python
#group by studio
movies_and_studios.group('Studio')
```

```python
#sort and draw chart
movies_and_studios.group('Studio').sort('count', descending=True).barh(0)
```

```python
#I wanted to create a new table just with title and total gross in millions, here is the way
millions = top.select(0).with_columns('Gross - Mil Dollars', np.round(top.column(3)/1000000,2))
millions
```

```python
# histogram
millions.hist('Gross - Mil Dollars', unit='$m')
```

```python
millions.hist('Gross - Mil Dollars', bins=np.arange(300,2001,100), unit='$m')
```

```python
# Read the full Census table
census_url = 'http://www2.census.gov/programs-surveys/popest/datasets/2010-2015/national/asrh/nc-est2015-agesex-res.csv'
full_census_table = Table.read_table(census_url)

# Select columns from the full table and relabel some of them
partial_census_table = full_census_table.select(['SEX', 'AGE', 'POPESTIMATE2010', 'POPESTIMATE2015'])
us_pop = partial_census_table.relabeled('POPESTIMATE2010', '2010').relabeled('POPESTIMATE2015', '2015')

# Access the rows corresponding to all children, ages 0-18
children = us_pop.where('SEX', are.equal_to(0)).where('AGE', are.below(19)).drop('SEX')
children.show()
```

```python
children.plot('AGE')
```
